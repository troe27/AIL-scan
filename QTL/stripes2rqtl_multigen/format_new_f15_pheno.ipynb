```python
pedigree_new = {}
for i, k in a.iterrows():
    ID = str(k["ID"])+"15"
    if not ID in ped.pedigree:
        pedigree_new[ID]={}
        pedigree_new[ID]["Hatch"] = k["Hatch"]
        try:
            ped.pedigree[str(k["Sire"])+"14"]
            pedigree_new[ID]["Sire"] = str(k["Sire"])+"14"
        except KeyError:
            print(ID)
            print("the sire {} is not in the old pedigree".format( str(k["Sire"])+"14"))    
        try:
            ped.pedigree[str(k["Dam"])+"14"]
            pedigree_new[ID]["Dam"] = str(k["Dam"])+"14"
        except KeyError:
            print("the dam {} is not in the old pedigree".format( str(k["Dam"])+"14"))
        if k["Sex"] in ["F", "f"]:
            pedigree_new[ID]["Sex"]= 2
        elif k["Sex"] in ["M", "m"]:
            pedigree_new[ID]["Sex"]= 1
        else:
            pedigree_new[ID]["Sex"]= np.nan
    
    
    
```
