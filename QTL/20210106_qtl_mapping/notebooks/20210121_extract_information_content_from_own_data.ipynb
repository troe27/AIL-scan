```python
def translate_AA(x):
    if x ==1:
        y=1
    elif x ==0:
        y=0
    elif x ==-1:
        y=0
    elif x == 'NA':
        y=0
    else:
        print(x)
        raise ValueError('value is not either 1,0,-1 or np.nan')
    return y

def translate_AB(x):
    if x ==1:
        y=0
    elif x ==0:
        y=1
    elif x ==-1:
        y=0
    elif x == 'NA':
        y=0.5
    else:
        raise ValueError('value is not either 1,0,-1 or np.nan')
    return y

def translate_BB(x):
    if x ==1:
        y=0
    elif x ==0:
        y=0
    elif x ==-1:
        y=1
    elif x == 'NA':
        y=0
    else:
        print(x)
        raise ValueError('value is not either 1,0,-1 or np.nan')
    return y
```

```python
WBgeno = pd.read_csv('../data/Wahlberg_genotype_prob/hl070831.khcoeffqe', sep='\t', header=None)
WBgeno.columns = ['chrom', 'individual', 'cM', 'AA', 'AB', 'BA', 'BB']
wb_a = WBgeno['AA'] - WBgeno['BB']
WBgeno['a'] = wb_a
wb_d = WBgeno['AB']+WBgeno['BA']
WBgeno['d'] = wb_d
WBgeno_chrom = WBgeno.groupby('chrom')
```

```python

aadict = {}
dddict = {}

for i,k in WBgeno_chrom:
    k_cM = k.groupby('cM')
    for j, l in k_cM:
        for m,n in l.iterrows():
            #print(i,j, n['a'])
            aadict.setdefault(str(i)+'-'+str(j),{})[n['individual']]=n['a']
            dddict.setdefault(str(i)+'-'+str(j),{})[n['individual']]=n['d']
aa = pd.DataFrame(aadict)
dd = pd.DataFrame(dddict)
a_var = np.var(aa, axis=0)
d_var = np.var(dd)

```

```python
wb_inform = (a_var+(2*d_var))
wb_max = wb_inform.rolling(3).max()
wb_slide= np.convolve(wb_inform,np.ones(windowsize),"valid")/windowsize
```
