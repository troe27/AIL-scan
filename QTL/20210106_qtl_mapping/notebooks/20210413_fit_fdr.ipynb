```r
chr <- fdrpeaks$chr
pos <- fdrpeaks$pos
lodcol <- fdrpeaks$lod
```

```r
dt <-data.table(chr)
dt$pos <- pos
dt$lod <- lodcol
dt$mrk <- fdrpeaks$marker
```

```r
chrfdr <- dt[dt$chr==4]$chr
```

```r
chrfdr <- dt$chr
posfdr <- dt$pos
qtl_fdr <- makeqtl(datf2.5, chr = chrfdr, pos = posfdr, what = 'prob' )
fit_fdr <- fitqtl(datf2.5, pheno.col = 5, qtl = qtl_fdr, get.ests=TRUE)
summary(fit_fdr)
```

```r
chrom <- chrfdr
pos <- posfdr
mrk <- dt$mrk
for (qtl in seq(1,length(chrom))){
    curr_chrom <- chrom[qtl]
    curr_pos <- pos[qtl]
    print(mrk[qtl])
    qtls <- makeqtl(datf2.5, chr = curr_chrom, pos = curr_pos, what = 'prob')
    fit_sugg <- fitqtl(datf2.5, pheno.col = 5, qtl = qtls,  get.ests=TRUE, method = 'hk')
    print(summary(fit_sugg))
}
```
