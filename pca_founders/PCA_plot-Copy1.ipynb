```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from cyvcf2 import VCF
```

```python
samples = []
with open("./pca_data2.tsv") as handle:
    for line in handle.read().split("\n"):
        l = line.split("\t")
        #print l
        try:
            lnum = [float(i) for i in l]
            samples.append(lnum)
        except ValueError:
            print(l)
        
```

```python
with open("../find_fixed_sites/high_full.txt", "r") as handle:
    high_dict = dict.fromkeys([ i for i in handle.read().rsplit("\n") if not i ==""])
```

```python
name_dict = dict.fromkeys(names)
for key in list(high_dict.keys())+list(low_dict.keys()):
    if key in name_dict:
        print(key)
    else: 
        print(key, "WARNING")
```

```python
col =[]
for i in names:
    c = "purple"
    if i in high_dict:
        c = "red"
        print(i, "high")
    elif i in low_dict:
        c = "blue"
        print(i, "low")
    else:
        print(i, "UNSPECIFIED")
    col.append(c)
```

```python
#print col
import matplotlib.patches as mpatches


red_patch = mpatches.Patch(color='red', label='high')
blue_patch = mpatches.Patch(color='blue', label='low')
purple_patch = mpatches.Patch(color='purple', label='unspecified')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(30,10), sharex=False, sharey=False)
ax1.scatter(s_df[0],s_df[1], color=col, marker="o")
ax1.set_ylabel("PC_2 (0.02714638)", size=20)
ax1.set_xlabel("PC_1 (0.45680207)", size=20)
ax1.legend(handles=[red_patch, blue_patch, purple_patch],loc=0 )

ax2.scatter(s_df[0],s_df[2], color=col, marker="o")
ax2.set_ylabel("PC_3 (0.02134499)", size=20)
ax2.set_xlabel("PC_1 (0.45680207)", size=20)
ax2.legend(handles=[red_patch, blue_patch, purple_patch],loc=0 )

ax3.scatter(s_df[1],s_df[2], color=col, marker="o")
ax3.set_ylabel("PC_3 (0.02134499)", size=20)
ax3.set_xlabel("PC_2 (0.02714638)", size=20)
ax3.legend(handles=[red_patch, blue_patch, purple_patch],loc=0 )




plt.tight_layout()
plt.savefig("/Users/tilro532/wd/chicken_finemapping/pca_founders/pca_founders_1v2v3")
plt.show()
```
